# Excercise 2

```python
x = var('x')
```

```python
l = var('l')
```

```python
f = l * x * (1 - x)
```

```python
f.subs(l=1)
```

```python
def plot_f_for(f, __x0, __l):
    points = []
    x0 = __x0
    for i in range(0, 200):
        points.append([x0, f.subs(l=1, x = x0)])
        x0 = f.subs(l=__l, x = x0)
    plot = None
    for i in range(0, len(points)):
        tmp = line([i, points[i]])
        if plot is None:
            plot = tmp
        else:
            plot += tmp
    return plot
```

```python
plot_f_for(f, 0, 1).show()
```

```python
plot_f_for(f, 0.5, 1).show()
```

```python
plot_f_for(f, 0.7, 1).show()
```

```python
plot_f_for(f, 0.5, 2).show()
```

```python
plot_f_for(f, 0.5, 3.6).show()
```

# Excercise 3

```python
def get_value_after_iterations(f, __x0, it):
    x0 = __x0
    for i in range(0, it):
        x0 = f.subs(x = x0)
    return x0
```

```python
f2 = 4*x*(1-x)
```

```python
get_value_after_iterations(f2, 0.67, 40)
```

```python
get_value_after_iterations(f2, 0.67001, 40)
```
