```python
x = var('x')
y = var('y')
```

```python
f1 = -y+x*(1-x**2-y**2)
f2 = x+y*(1-x**2-y**2)
```

```python
plot_vector_field([f1,f2],(x,-2,2),(y,-2,2))
```

```python
times = srange(0, 6, 0.1)
the_plot = plot_vector_field([f1,f2],(x,-3,3),(y,-3,3))
for ics in [(0.1, 0.1), (0.1, 0.5), (2.5, 0.1), (2.5, 0.5), (2, 2)]:
    sol = desolve_odeint([f1,f2], ics, times, [x, y])
    the_plot += line(zip(sol[:,0], sol[:, 1]))
show(the_plot)
```
